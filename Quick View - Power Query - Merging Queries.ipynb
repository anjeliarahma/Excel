# **Power Query - Merging Queries**

# Scheme

Screenshot 2024-11-01 215611.png

**1. Inner Join**
Is the table produced from the join contains rows for each pair of rows from the specified tables that match based on the specified key columns.

**2. Left Outer Join**
Is a left outer join ensures that all rows from the first table appear in the result.

**3. Left Anti Join**
Is a left anti join returns all rows from the first table that do not have a match in the second table.

**4. Right Anti Join**
Is a  right anti join returns all rows from the second table that do not have a match in the first table.

**5. Right Outer Join**
Is a right outer join ensures that all rows from the second table appear in the result.

**6. Full Outer Join**
Is a full outer join ensures that all rows from both tables appear in the result. Rows that do not have a match in the other table are combined with default rows containing null values for all their columns.
